```python
import numpy as np
import pandas as pd
```

```python
data = np.random.randint(0, 10, 5)
index = ['a','b','c','d','e']
series = pd.Series(data=data, index=index, name='series_from_ndarray')
print(series)
```

```python
data = {'서울' : 70000, '부산' : 52000, '대전' : 35000, '광주' : 40000, '제주도' : 28000}  # 딕셔너리
series = pd.Series(data=data)   # index는 key 값으로 대체가 된다.
print(series)
```

```python
city = ['광주', '대전', '부산', '서울', '울릉도']
series = pd.Series(data=data, index=city, name='series_from_dict')   # 순서 바꾸기 가능, 울릉도는 결측값
print(series)
```
