```python
data_1 = [5, 9, 2.5, 3, 1]
array_1 = num.array(data_1)    # 순차형 객체를 넘파이 배열로.
print(array_1)

data_2 = [[1, 3, 5, 7, 9],[2, 4, 6, 8, 10]]   # 같은 길이의 리스트를 포함한 순차 데이터 -> 동일한 다차원 배열로
array_2 = num.array(data_2)
print(array_2)

print(array_2.ndim)    # 차원
print(array_2.shape)   # 차원의 크기
print(array_1.dtype)   # 데이터 타입. 다른 게 나옴.
print(array_2.dtype)
```

```python
array_1 = num.array([1,3,5,7,9], dtype=num.float32)   # type 지정
array_2 = num.array([1,3,5,7.9], dtype=num.int16)

print(array_1.dtype)
print(array_2.dtype)
```

```python
array_2_float = array_2.astype(num.float64)   # astype으로 타입 바꿀 수 있음.
array_2_float.dtype
```
