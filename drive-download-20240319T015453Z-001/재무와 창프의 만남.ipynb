```python
class Equity(object) : # object는 걍 넣으면 되는 듯?
  def __init__(self, D='', P='', r='', t='', g='') :
    self.dividend = D
    self.price = P
    self.r_return = r
    self.period = t
    self.growth = g

  def get_dividend(self) :
    return self.dividend
  def get_price(self) :
    return self.price
  def get_r_return(self) :
    return self.r_return
  def get_period(self) :
    return self.period
  def get_growth(self) :
    return self.growth

  def set_dividend(self, newD) :
    self.dividend = newD
  def set_price(self, newP) :
    self.price = newP
  def set_r_return(self, newr) :
    self.r_return = newr
  def set_period(self, newt) :
    self.period = newt
  def set_growth(self, newg) :
    self.growth = newg
```

```python
samsung = Equity(2000, 60000, 0.05, t='', g=0.1)
```

```python
samsung.get_dividend()
```
