```python
# 칼럼 순서 변경
df = df[['펭귄의 종류','서식하는 섬','성별','부리 길이','부리 깊이','지느러미 길이','몸무게']]
df.head()
```

```python
# for문과 if문으로 칼럼을 이동시킬 수 있다고
df=sns.load_dataset('penguins')
cols_to_stay = ['bill_length_mm','bill_depth_mm','flipper_length_mm']
cols = cols_to_stay + [col for col in df if col not in cols_to_stay]    # 리스트 만든 건데. 읽으면 이해 될 거임.
df = df[cols]                                                           # df = df[[내용]] 이거로 칼럼 순서 바꾼거.
df.head()
```

```python
# 칼럼과 인덱스 교환
df = df[:11].transpose()   # 길어서 11번까지만 가져옴.
df
```
