# 이상값 처리

```python
'''
이상값이 아무 의미도 없는 것은 아님. 데이터셋이 속한 도메인이나 종류에 따라 큰 의미가 있기도 하다.
도메인에 속한 전문가와 이상값을 검토하는 게 좋다. 이상값이 생긴 이유를 명확하게 아는 것이 좋기 때문이다.
'''
```

```python
# 이상값 확인하는 가장 쉽고 간편한 방법. 시각화
# 1개변수. 박스플롯. 히스토그램.   2개변수 산점도
```

```python
sns.set_theme(style='darkgrid')
plt.figure(figsize=(12,4))
sns.boxplot(x=titanic.Fare);
```

```python
plt.figure(figsize=(12,4))
sns.boxplot(x=titanic.Age, color='yellowgreen');
```

```python
''' IQR 기법   EDA 선구자 존 튜키 선생이 만들었다고
1. 1사분위수 Q1을 찾음
2. 3사분위수 Q3를 찾음
3. IQR 계산. = Q3-Q1

하한값 Q1-1.5*IQR  상한값 Q3+1.5*IQR
벗어난 값은 이상값으로 간주.'''
```

```python
def outlier_iqr(data,column) :
  global lower, upper   # 하한, 상한 : 글로벌 변수 선언
  q1, q3 = np.quantile(data[column], 0.25), np.quantile(data[column], 0.75)  # 튜플이죠 맞을걸. 4분위수 기준 지정
  iqr = q3 - q1   # iqr 계산
  cut_off = iqr * 1.5   # outlier cutoff 기준점 계산하기
  lower, upper = q1-cut_off, q3+cut_off
  print('IQR은 ',iqr, '이다.')
  print('lower 기준값은 ',lower, '이다.')
  print('upper 기준값은 ',upper, '이다.')

  data1 = data[data[column]>upper]
  data2 = data[data[column]<lower]

  return print('총 이상값 개수는 ', data1.shape[0]+data2.shape[0],' 이다.')
```

```python
outlier_iqr(titanic, 'Fare')   # 변수값에 -26은 없기에 무시해도 됨. 음수의 요금은 없음.
```
