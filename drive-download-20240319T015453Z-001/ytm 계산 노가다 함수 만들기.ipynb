```python
def func(pv, c, par, t, y) :
  a = c/(1+y)*(1-1/(1+y)**t) + par/(1+y)**t - pv
  return a
```

```python
func(9000, 1000, 10000, 3, 0.1)
```

```python
pv=9000
c=1000
par=10000
t=3
func(pv,c,par,t,c/par)
```

```python
def ytm(pv, c, par, t) :
  y=c/par
  while True :
    b= func(pv,c,par,t,y)
    print(y)
    if round(b) > 0 :
      y = y+0.001
    elif round(b) < 0 :
      y = y-0.001
    elif round(b) ==0 :
      break
  return print('찾으시는 값은 ',y,'입니다.')
```

```python
ytm(900, 10, 1000, 3)
```

```python
func(900, 10, 1000, 3,0.036)
```

```python
ytm(600, 10, 1000, 5)
```

```python
# 제가 만든 노가다 함수를 IRR 계산에도 사용해 보겠습니다.
# 문제가 있다면 일정한 cash flow가 아니네여
```

```python
def npv(i, c, t, y) : # t가 3년과 4년만 되는 npv 계산기
  if t == 3 :
    return c[0]/(1+y)**1 + c[1]/(1+y)**2 + c[2]/(1+y)**3 - i

  else :
    return c[0]/(1+y)**1 + c[1]/(1+y)**2 + c[2]/(1+y)**3 + c[3]/(1+y)**4 - i
```

```python
A = [17300, 16200, 13800, 7600]
B = [5700, 12900, 16300, 27500]

NPVa = npv(37500, A, 4, 0.11)
NPVb = npv(37500, B, 4, 0.11)

print(NPVa)
print(NPVb)
```

```python
C = print([Ai - Bi for Ai, Bi in zip(A, B)])
print(C)
```

```python
def IRR(i, c, t) : # t는 4라는 전제, NPV 식에 t를 넣어서 어쩔 수 없네
  if i/abs(c[0]+c[1]+c[2]+c[3]) > 1 :
    y = i/abs(c[0]+c[1]+c[2]+c[3])
  else :
    y = abs(c[0]+c[1]+c[2]+c[3]) / i # 이러면 마이너스 아닌가. 아 모르겠다. 대충 만드는 건데 뭘 그렇게 신경쓰냐.

  while True :
    b= NPV(i, t, c, y)
    print(y)
    if round(b) > 0 :
      y = y+0.001
    elif round(b) < 0 :
      y = y-0.001
    elif round(b) ==0 :
      break
  return print('찾으시는 값은 ',y,'입니다.')
```
